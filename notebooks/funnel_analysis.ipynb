1.Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Libraries installed successfully!")
```

2.Load Dataset

```python
df = pd.read_csv("C:\\FUTURE_DS_03\\data\\olist_closed_deals_dataset.csv")

print("Dataset Shape:", df.shape)

df.head()
```

```python
print("Size of dataset:", df.size)
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(df.columns)
```

3.Data Cleaning

```python
df.drop_duplicates(inplace=True)

df['business_segment'] = df['business_segment'].fillna('Unknown')
df['lead_type'] = df['lead_type'].fillna('Unknown')
df['lead_behaviour_profile'] = df['lead_behaviour_profile'].fillna('Unknown')

print("Dataset Shape:", df.shape)
```

4.Convert Date Column

```python
df['won_date'] = pd.to_datetime(df['won_date'])

df.head()
```

5.Total Closed Deals

```python
total_deals = len(df)

print("Total Closed Deals:", total_deals)
```

6.Business Segment Analysis

```python
business_segment = (
    df['business_segment']
    .value_counts()
    .head(10)
)

business_segment
```

7.Business Segment Chart

```python
plt.figure(figsize=(10,5))

business_segment.plot(kind='bar')

plt.title("Top Business Segments")
plt.xlabel("Business Segment")
plt.ylabel("Number of Deals")

plt.show()
```

8.SDR Performance

```python
sdr_performance = (
    df['sdr_id']
    .value_counts()
    .head(10)
)

sdr_performance
```

9.SDR Performance Chart

```python
plt.figure(figsize=(10,5))

sdr_performance.plot(kind='bar')

plt.title("Top 10 SDR Performance")
plt.xlabel("SDR ID")
plt.ylabel("Deals Closed")

plt.show()
```

10.Sales Representative Performance

```python
sr_performance = (
    df['sr_id']
    .value_counts()
    .head(10)
)

sr_performance
```

11.Sales Representative Chart

```python
plt.figure(figsize=(10,5))

sr_performance.plot(kind='bar')

plt.title("Top Sales Representatives")
plt.xlabel("SR ID")
plt.ylabel("Deals Closed")

plt.show()
```

12.Company Analysis

```python
df['has_company'] = df['has_company'].fillna(False)

df['has_company'].value_counts()
```

13.Company Distribution Chart

```python
df['has_company'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Has Company Distribution")
plt.ylabel("")

plt.show()
```

14.Monthly Deal Trend

```python
df['month'] = df['won_date'].dt.to_period('M')

monthly_deals = (
    df.groupby('month')
    .size()
)

monthly_deals
```

15.Monthly Deal Trend Chart

```python
plt.figure(figsize=(12,5))

monthly_deals.plot(marker='o')

plt.title("Monthly Closed Deals Trend")
plt.xlabel("Month")
plt.ylabel("Deals")

plt.grid(True)

plt.show()
```

16.Save Processed Dataset

```python
df.to_csv(
    "C:\\FUTURE_DS_03\\data\\processed_marketing_data.csv",
    index=False
)

print("Processed dataset saved successfully.")
```
